```python
import pandas as pd

df_clean = pd.read_csv("../data/carbon_clean.csv")
```

```python
df_world = df_clean[df_clean["country"] == "World"]

df_country = df_clean[
    (df_clean["iso_code"].notna()) &
    (df_clean["iso_code"].str.len() == 3)
]
```

```python
latest_year = df_country["year"].max()

latest = df_country[df_country["year"] == latest_year]
```

```python
world_2024 = df_world[df_world["year"] == latest_year]

world_2024["co2"].iloc[0]
```

```python
global_co2 = world_2024["co2"].iloc[0]

print("Global CO2:", global_co2)
```

```python
largest_emitter = latest.nlargest(1, "co2")

print(largest_emitter[["country", "co2"]])
```

```python
highest_per_capita = latest.nlargest(1, "co2_per_capita")

print(highest_per_capita[["country", "co2_per_capita"]])
```

```python
highest_gdp = latest.nlargest(1, "gdp")

print(highest_gdp[["country", "gdp"]])
```

```python
top_emitters = (
    latest
    .sort_values("co2", ascending=False)
    [["country", "co2"]]
    .head(10)
)

top_emitters
```

```python
top_emitters.to_csv("../data/top_emitters.csv", index=False)

latest.to_csv("../data/latest_country_data.csv", index=False)
```
